```python
import pandas as pd

# create a dataframe by reading CSV data from a URL

data_url = 'https://nes-lter-data.whoi.edu/api/ctd/en617/metadata.csv'

casts = pd.read_csv(data_url, parse_dates=['date'], index_col='cast')
casts.head(5) # "head" means just show the first 5 rows
```

```python
# getting the unique values in a column
casts['nearest_station'].unique()
```

```python
# logical indexing on equality
casts[casts['nearest_station'] == 'L11']
```

```python
# logical indexing: inequality
casts[casts['latitude'] > 41]
```

```python
# now let's read some data about sampling on the casts

URL = 'https://nes-lter-data.whoi.edu/api/ctd/en617/bottle_summary.csv'
bottles = pd.read_csv(URL, parse_dates=['date'])
bottles.head()
```

```python
bottles[bottles['cast'] == 1]
```

```python
# how many bottles per cast?
group = bottles.groupby('cast')
group['niskin'].count()
```

```python
# what's the maximum sampling depth per-cast?
max_depth = group['depth'].max()
max_depth
```

We can merge the max depth Series into the dataframe by using the `merge` function, which aligns on index values:

```python
# now add that to the existing casts dataframe
merged = casts.merge(max_depth, left_index=True, right_index=True)
merged.head()
```

```python
%matplotlib inline

merged.plot.scatter(x='latitude', y='depth');
```
